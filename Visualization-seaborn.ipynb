```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(color_codes=True, style='darkgrid')
```

```python
import pandas as pd
df = pd.read_csv('data/winequality-red.csv', delimiter=';')
```

```python
# Revisit the data to see what it looks like. 
df.head()
df.describe()
```

```python
# Let's try a histogram first. 

# Without kernel density estimation plot
sns.distplot(df['quality'], bins=6, kde=False, vertical=False, axlabel='Quality')
plt.ylabel('Frequency')
plt.title('Histogram of quality column')
plt.show()
```

```python
# With kernel density estimation
#plt.figure()
p = sns.distplot(df['quality'], bins=6, kde=True, vertical=False, axlabel='Quality')
p.set_ylabel('Frequency')
p.set_title('Histogram of quality column')
plt.show()
```
